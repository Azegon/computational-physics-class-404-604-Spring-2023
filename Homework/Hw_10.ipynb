\begin{equation}
\nabla^2\Phi=4\pi G\rho
\end{equation}
with 
\begin{equation}
\nabla^2\Phi=\frac{\partial^2\Phi}{\partial^2 x}+\frac{\partial^2\Phi}{\partial^2y}
\end{equation}

\begin{equation}
4\pi G\rho=\frac{\partial^2\Phi}{\partial^2 x}+\frac{\partial^2\Phi}{\partial^2y}
\end{equation}
With G = 1

```python
import matplotlib.pylab as p;
from mpl_toolkits.mplot3d import Axes3D
from numpy import *;
import numpy;
print("Initializing")

Nmax = 100; Niter = 5000; V = zeros((Nmax, Nmax), float)   

#boundry conditions
#central density condition
G = 1
roe = 1
V[50,50] = -4*numpy.pi*G*roe
    
for iter in range(Niter):                                  
    if iter%10 == 0: print(iter)
    maxerr=0.
    for i in range(1, Nmax-1):                                                
        for j in range(1,Nmax-1): 
            #condition that Roe[50,50] = 1, skips changing point (50,50)
            if (i == 50 & j == 50):
                continue
            #regular condition
            du = 1./4.*(V[i+1,j]+V[i-1,j]+V[i,j+1]+V[i,j-1])-V[i,j] #
            if(maxerr<abs(du)): # how much correction we have in this step
                maxerr=abs(du)  #
            V[i,j]=  V[i,j] + du# please finish this line for Gauss-Siedel method
    #print(maxerr)
    if(maxerr<1.e-3):
        break
            
x = range(0, Nmax, 1)
y = range(0, Nmax, 1)                              
X, Y = p.meshgrid(x,y)                 

def functz(V):                                         # V(x, y) 
    z = V[X,Y]                        
    return z

Z = functz(V)                          
fig = p.figure()  
ax = fig.add_subplot(projection='3d')                                        # Plot axes
ax.plot_wireframe(X, Y, Z, color = 'r')               # Red wireframe
ax.set_xlabel('X')                                     
ax.set_ylabel('Y')
ax.set_zlabel('Potential')
p.show()                                              # Show fig
```
