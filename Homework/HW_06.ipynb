```python
import numpy as np
import matplotlib.pyplot as plt

#write rutta kada method rk4
#constants and starting values
g = 9.81
y = [0, 10]
tfin = 2.1
t = 0
h = 0.1

#equation  vector
def f(t, y):
    f=np.zeros(2)
    #dx/dt
    f[0] = y[1]*t
    #dV_y/dt
    f[1] = - g*t

    return f

def rk4(t, h, y, n):
    k1 = np.zeros(n)
    k2 = np.zeros(n)
    k3 = np.zeros(n)
    k4 = np.zeros(n)
    
    k1 = h*f(t,y)
    k2 = h*f(t+0.5*h, y+0.5*k1)
    k3 = h*f(t+0.5*h, y+0.5*k2)
    k4 = h*f(t+h, y+k3)
    y = y + (k1 + 2*k2 + 2*k3 + k4)/6
    print(y)
    return y

tp = []
yp = []
while (t<tfin):
    if ((t+h)> tfin):
        h = tfin - t
    y = rk4(t, h, y, 2)
    yp = np.append(yp, y[0])
    t = t+h
    tp = np.append(tp,t)
    
plt.plot(tp,yp)
plt.show()
```
