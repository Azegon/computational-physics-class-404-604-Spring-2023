```python
#Tabris Loveless
#Phys 404
#2/1/2023

import numpy as np
import matplotlib.pyplot as plt

#part 1 use memorization method to calculate and plot fib(n)/fib(n-1)

def fibMem(n, table):
    """
    Calculates the fibonacci numbers using memorization method

    Args:
    n: int
        term of fibonacci sequence.
    table: list
        list of already calculated fibonacci numbers
    """
    #end condition
    if n == 0 or n == 1:
        table[n] = n
        
        #recursion call
    if table[n] is None:
        table[n] = fibMem(n-1, table) + fibMem(n-2, table)
        
    return table[n]

def graphFibRatio(end):
    """
    Graphs the ratio of fib(n)/fib(n-1) from n = 2 to n = end
    
    args:
        end: int
        highest nth term of graph
        !numbers over 2000 can exheed recursion depth
    """
    #table stores calculated fib numbers
    table = [None]*(end+1)
    
    #calculation of fib number to end term
    print(fibMem(end, table))
    
    #calulate f(n)/f(n-1) for n = 2 to n = end
    x = range(2, end+1)
    y = []
    
    for ix in x:
        ratio = table[ix]/table[ix-1]
        y.append(ratio)
    
    plt.plot(x, y)
    plt.title('Graph of fib(n)/fib(n-1)')
    plt.xlabel('x')
    plt.ylabel('Fib(n)/Fib(n-1)')
    plt.show()
```

```python
#part 2
def fibRec(n):
    #recursion method
    #limited n values
    if n == 0: return 0
    elif n == 1: return 1
    else: return fibRec(n-1)+fibRec(n-2)
    
def fibTab(n):
    #tabulation method 
    table=[0]*(n+1)
    table[1] = 1
       
    for ix in range(2,n+1):
        table[ix] = table[ix-1]+table[ix-2]
    
    return table[n]


def arrpow(arr, n):
    yarr=arr
    if n<1:
        print('n needs to be larger than 1')
    if n==1:
        return arr
    yarr = arrpow(arr, n//2)
    yarr = [[yarr[0][0]*yarr[0][0]+yarr[0][1]*yarr[1][0],yarr[0][0]*yarr[0][1]+yarr[0][1]*yarr[1][1]],
            [yarr[1][0]*yarr[0][0]+yarr[1][1]*yarr[1][0],yarr[1][0]*yarr[0][1]+yarr[1][1]*yarr[1][1]]]
    if n%2:
        yarr=[[yarr[0][0]*arr[0][0]+yarr[0][1]*arr[1][0],yarr[0][0]*arr[0][1]+yarr[0][1]*arr[1][1]],
            [yarr[1][0]*arr[0][0]+yarr[1][1]*arr[1][0],yarr[1][0]*arr[0][1]+yarr[1][1]*arr[1][1]]]
    return yarr

def fib5(n):
    arr= [[1,1],[1,0]]
    f=arrpow(arr,n-1)[0][0]
    return f
    
```

```python
import time
import csv

n_values = range(1, 41)
#recursion method test
rec_time = []
for ix in n_values:
    start = time.time()
    temp_fib = fibRec(ix)
    end = time.time()-start
    rec_time.append(end)
    print(rec_time[-1], end="\r")

#make file
with open('Recursion.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(n_values)
    writer.writerow(rec_time)
    
print('File written')
```

```python
tab_values = range(1, 100000, 100)
#tabulation method
tab_time = []
for ix in tab_values:
    start = time.time()
    temp_fib = fibTab(ix)
    end = time.time()-start
    tab_time.append(end)
    print(tab_time[-1], end="\r")

with open('Tabulation.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(tab_values)
    writer.writerow(tab_time)
    
print('File written')
```
