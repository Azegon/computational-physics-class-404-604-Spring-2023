```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as p
from mpl_toolkits.mplot3d import Axes3D


def burgerPlot(N):
    """
    Function solves and plots Burger's equation for given grid space size N.
    
    Args:
        N int:
        Size of grid space.
    """
    #init parameters
    xmin = -10.0   # left boundary
    xmax = +10.0   # right boundary
    tend = 10.0    # end time
    dtout = 1.0   # time interval for outputs
    
    #grid space
    x = np.linspace(xmin, xmax, N+1)
    dx = (xmax - xmin) / (N)
    dt = 0.8 * dx
    U = np.zeros(N+1,)
    
    #init profile
    t = 0.0
    U = 0.2 + 0.8 * np.exp(-0.5 * x**2)

    #3d output
    tp = [t]
    Up = np.copy(U)
    
    #3d plot init
    fig = plt.figure(figsize=(16,10))
    ax = fig.gca()
    ax.plot(x, U, label=r"$t = {:.3G}$".format(t))
    ax.set_xlabel(r"Position $x$")
    ax.set_ylabel(r"$U(t,x)$")
    ax.minorticks_on()
    
    #calculation
    tout = t + dtout
    while t < tend:
        # Backup the previous time step.
        Uold = np.copy(U)

        for ix in range(1, N):
            U[ix]=Uold[ix] - (dt/dx)*Uold[ix]*(Uold[ix] - Uold[ix-1])# Task: implement upwind method here.
            
        t += dt

        # Save the data after every dtout.
        if t >= tout:
            plt.plot(x, U, label=r"$t = {:.3G}$".format(t))
            tp.append(t)
            Up = np.vstack((Up, U))
            print("t = ", t)
            tout += dtout

    ax.legend()

    #plotting
    t, x = np.meshgrid(tp, x)
    fig3D = plt.figure(figsize=(16,10))
    ax3D = Axes3D(fig3D)
    ax3D.plot_wireframe(t, x, Up.transpose(), color="red")
    ax3D.set_xlabel("Time $t$")
    ax3D.set_ylabel("Position $x$")
    ax3D.set_zlabel(r"$U(t,x)$")

    print("Done.")
    plt.show()
```

```python
burgerPlot(100)
```

```python
#modified grid space
burgerPlot(200)
```

```python
burgerPlot(500)
```

```python
burgerPlot(1000)
```

Across the three plots, we see a exponential decay of the wave over time. This exponential decay is delayed on the larger grid space calculations.
